### Spliting the Data

```python
X=dsn.drop('W',axis=1)
y=dsn['W']
```

### Machine Learing Model

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Define a function to try different train test folds
def model_select(model):
    max_score=0
    max_state=0
    for x in range(42,105):
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=x)
        md=model()
        md.fit(X_train,y_train)
        predict=md.predict(X_test)
        r2score=r2_score(y_test,predict)
        if r2score>max_score:
            max_score=r2score
            max_state=x
    print("Max score is {} with random state {}".format(max_score,max_state))
```

```python
model_select(LinearRegression)
```

```python
model_select(ElasticNet)
```

```python
model_select(Ridge)
```

```python
model_select(KNeighborsRegressor)
```

```python
model_select(RandomForestRegressor)
```

```python
model_select(AdaBoostRegressor)
```

```python
model_select(DecisionTreeRegressor)
```

```python
model_select(Lasso)
```

### Lasso Regression best model

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.33,random_state=50)
```

```python
le=Lasso()
le.fit(x_train,y_train)
y_pred=le.predict(x_test)
print('mse ',mean_squared_error(y_test,y_pred))
print('rmse ',mean_squared_error(y_test,y_pred,squared=False))
print('mean absolute error ',mean_absolute_error(y_test,y_pred))
print('r2 score ',r2_score(y_test,y_pred))
```

```python
mse=cross_val_score(l,X,y,scoring='neg_mean_squared_error',cv=7)
print(-mse)
print('mean ',-mse.mean())
```

lets do some hyperparameter tuning to improve the model accuracy

Now lets train the model with complete data set

### Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_r ={'alpha':[0.1,0.01,0.001,0.0001],
       'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']}
```

```python
ridge_hp = GridSearchCV(Ridge(),param_r,cv=5)
ridge_hp.fit(x_train,y_train)
ridge_hp.best_params_
```

```python
# Ridge model
rg =Ridge(alpha=0.1,solver ='saga')
rg.fit(x_train,y_train)
rg.score(x_train,y_train)
predict = rg.predict(x_test)

print(r2_score(y_test,predict))
```

```python
# Lasso Regrssion param

param ={'alpha':[0.1,0.01,0.001,0.0001],
       'selection':['cyclic', 'random']}
```

```python
lasso_hp = GridSearchCV(Lasso(),param,cv=5)
lasso_hp.fit(x_train,y_train)
lasso_hp.best_params_
```

```python
# Through alpha value it will be corrected
le =Lasso(alpha=0.1,selection='random')
le.fit(x_train,y_train)
predict = le.predict(x_test)
le.score(x_train,y_train)

print(r2_score(y_test,predict))
```

Lasso showed the best results

### Prediction
